We'll read in our csv with lead level sampling data from across the U.S. in 2018 and clean it.

```python
levelsdf = pd.read_csv('water_data_2018.csv')
levelsdf.rename(columns={'ActivityLocation/LatitudeMeasure': 'ActivityLatitude',
                   'ActivityLocation/LongitudeMeasure': 'ActivityLongitude',
                   'CharacteristicName':'Contaminant',
                   'ResultMeasureValue':'DetectedLevel',
                   'ResultMeasure/MeasureUnitCode':'LevelUnit'}, inplace=True)
levelsdf = pd.DataFrame(levelsdf[['ActivityIdentifier',
                                  'MonitoringLocationIdentifier',
                                  'AnalysisStartDate',
                                  'Contaminant',
                                  'DetectedLevel',
                                  'LevelUnit',
                                  'ActivityMediaSubdivisionName']])

#covert one-off string values to numerical values
levelsdf = levelsdf.replace('ND', 0)
levelsdf = levelsdf.replace('*Non-detect', 0)
levelsdf = levelsdf[levelsdf['DetectedLevel'] !='Not Reported']
levelsdf = levelsdf[levelsdf['DetectedLevel'] !='*Not Reported']

#convert rows with > or < on measured value
count = 1
for value in levelsdf['DetectedLevel']:
    if '<' in str(value):
        replacement = float((value.split('<')[1]))/2
        levelsdf.replace(value, replacement, inplace=True)
        print(f'replaced {count}')
        count += 1
        
count = 1
for value in levelsdf['DetectedLevel']:
    if '>' in str(value):
        replacement = float((value.split('>')[1]))
        levelsdf.replace(value, replacement, inplace=True)
        print(f'replaced {count}')
        count += 1
        
for value in levelsdf['DetectedLevel']:
    try:
        float(value)
    except:
        levelsdf = levelsdf[levelsdf['DetectedLevel'] !=value]
        print('dropped')

#remove rows with no value for detected levels
levelsdf = levelsdf[pd.notnull(levelsdf['DetectedLevel'])]

#covert detected values to floats
levelsdf['DetectedLevel'] = levelsdf.DetectedLevel.astype(float)

#remove rows with no value for detected levels after conversion
levelsdf = levelsdf[pd.notnull(levelsdf['DetectedLevel'])]

#change columns with unreadable data types to None for insertion into mySQL
levelsdf = levelsdf.where((pd.notnull(levelsdf)), None)

levelsdf = levelsdf.reset_index()

#change mg/L to ug/L
for i in list(range(len(levelsdf))):
    if levelsdf['LevelUnit'][i] == 'mg/l':
        levelsdf['DetectedLevel'][i] = levelsdf['DetectedLevel'][i] * 1000
        levelsdf['LevelUnit'][i] = 'ug/l'
        print('converted')
        
levelsdf = levelsdf[levelsdf['LevelUnit'] =='ug/l']

print('csv read')

```

Now we'll read in location data

```python
site_df = pd.read_csv('site_data_2018.csv')
site_df = pd.DataFrame(site_df[['MonitoringLocationIdentifier', 'LatitudeMeasure', 'LongitudeMeasure']])
```

And merge it with our sampling dataframe.

```python
lead_sites = levelsdf.merge(site_df, 
                            left_on = 'MonitoringLocationIdentifier', 
                            right_on = 'MonitoringLocationIdentifier')
```

```python
del lead_sites['index']
lead_sites = lead_sites.reset_index(drop=True)
```

```python
#create database
from mysql.connector import errorcode
db_name = 'water'

def create_database(cursor, database):
    try:
        cursor.execute(
            "CREATE DATABASE {} DEFAULT CHARACTER SET 'utf8'".format(database))
    except mysql.connector.Error as err:
        print("Failed creating database: {}".format(err))
        exit(1)

try:
    cursor.execute("USE {}".format(db_name))
except mysql.connector.Error as err:
    print("Database {} does not exist.".format(db_name))
    if err.errno == errorcode.ER_BAD_DB_ERROR:
        create_database(cursor, db_name)
        print("Database {} created successfully.".format(db_name))
        cnx.database = db_name
    else:
        print(err)
        exit(1)
```
